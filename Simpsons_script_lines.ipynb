```python
import nltk  
from nltk.tokenize import WordPunctTokenizer
#nltk.download()  
```

```python
#to lower case
Drop_false_speaking_line['spoken_words'] = Drop_false_speaking_line.spoken_words.map(lambda x: x.lower()) 

Drop_false_speaking_line
```

```python
#remove punctuations 
Drop_false_speaking_line['spoken_words'] = Drop_false_speaking_line.spoken_words.str.replace('[^\w\s]', '') 
Drop_false_speaking_line
```

```python
#tokenize into single words

from nltk.tokenize import word_tokenize
Drop_false_speaking_line['spoken_words'] = Drop_false_speaking_line['spoken_words'].apply(word_tokenize)

Drop_false_speaking_line
```

```python
#stemming
stemmer = PorterStemmer()

Drop_false_speaking_line['spoken_words'].apply(lambda x: [stemmer.stem(y) for y in x])
Drop_false_speaking_line['spoken_words'] = Drop_false_speaking_line['spoken_words'].apply(lambda x: [stemmer.stem(y) for y in x])
Drop_false_speaking_line
```

```python
from sklearn.feature_extraction.text import CountVectorizer
Drop_false_speaking_line['spoken_words'].apply(lambda x: ' '.join(x))
count_vect = CountVectorizer(max_features=1500, min_df=5, max_df=0.7, stop_words=stopwords.words('english'))
#type(count_vect)
Drop_false_speaking_line['spoken_words']=[" ".join(review) for review in Drop_false_speaking_line['spoken_words'].values]

#count_vect = CountVectorizer()  
#counts = count_vect.fit_transform(Drop_false_speaking_line['spoken_words'])  
```

```python
#tf-idf

counts = count_vect.fit_transform(Drop_false_speaking_line['spoken_words'])

transformer = TfidfTransformer().fit(counts)

counts = transformer.transform(counts)  
```

```python
from sklearn.model_selection import train_test_split
Drop_false_speaking_line_files = ['Drop_false_speaking_line']
Drop_false_speaking_line_files
#df = pd.DataFrame()


Drop_false_speaking_line_files = pd.read_csv('Drop_false_speaking_line',error_bad_lines=False,
                    warn_bad_lines=False,
                    low_memory=False)
X_train1,X_test1,y_train1,y_test1 = train_test_split(counts,
                                                 Drop_false_speaking_line_files["character_id"],
                                                 test_size=0.3,
                                                 random_state=int(5))

```

```python
#X_train1.info()
```

```python
#GaussianNB()
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB().fit(X_train1, y_train1)  
```

```python

predicted = model.predict(X_test1)

print(np.mean(predicted == y_test1))  
```
